```python
import matplotlib.pyplot as plt
import pandas as pd
```

1. MARGEM LÍQUIDA

```python
dados = {
    'Empresa': ['ALIANS', 'CCR', 'CYRELA', 'EZTEC', 'IGUATEMI', 'MULTIPLAN'],
    '2022': [19.99, 21.96, 15.86, 30.51, 1.26, 42.80],
    '2023': [128.82, 9.80, 17.68, 22.72, 26.70, 50.28],
    '2024': [29.64, 6.02, 24.12, 27.47, 32.34, 52.70],
    'Média': [59.48, 12.59, 19.22, 26.90, 20.10, 48.59]
}

df = pd.DataFrame(dados)

media_global = 23.50
```

```python
empresas = df['Empresa']
m2022 = df['2022']
m2023 = df['2023']
m2024 = df['2024']
media = df['Média']

# Configuração da figura  MESMA COISA DOS GRAFICOS DE ENERGIA
fig, ax = plt.subplots(figsize=(16, 8))
width = 0.25
x = range(len(empresas))

# Barras para cada ano
bars_2022 = ax.bar([p - width for p in x], m2022, width, label='2022', color='#66b3ff', edgecolor='black')
bars_2023 = ax.bar(x, m2023, width, label='2023', color='#ffcc99', edgecolor='black')
bars_2024 = ax.bar([p + width for p in x], m2024, width, label='2024', color='#99cc99', edgecolor='black')

# Linha da média individual por empresa
ax.plot(x, media, label='Média das Empresas', color='#ff6347', marker='o', linestyle='-', linewidth=3, markersize=8)

# Linha horizontal da média global do setor
ax.axhline(media_global, color='#ff4500', linestyle='--', label='Média Global (23,50%)', linewidth=2)

# Títulos e rótulos
ax.set_xlabel('Empresas', fontsize=16, fontweight='bold')
ax.set_ylabel('Margem Líquida (%)', fontsize=16, fontweight='bold')
ax.set_title('Margem Líquida das Empresas (2022-2024)', fontsize=18, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(empresas, rotation=45, ha='right', fontsize=12, fontweight='bold')

# Legenda principal
ax.legend(fontsize=12, loc='upper left', bbox_to_anchor=(0.0, -0.2), ncol=3)

# Criando legenda estendida com os valores por empresa/ano
numbers_legend = [
    f'{empresa}: {m2022[i]:.2f}% (2022), {m2023[i]:.2f}% (2023), {m2024[i]:.2f}% (2024)'
    for i, empresa in enumerate(empresas)
]

# Legenda extra com os números (no canto inferior direito)
fig.legend(numbers_legend, loc='lower right', fontsize=10, frameon=False, bbox_to_anchor=(1.05, 0), ncol=2)

# Layout final
plt.tight_layout()
plt.show()
```

NORMALIZAÇÃO DOS PESOS + RANKING

```python
dados = {
    'Empresa': ['ALIANS', 'CCR', 'CYRELA', 'EZTEC', 'IGUATEMI', 'MULTIPLAN'],
    'Liquidez Corrente': [2.30, 1.39, 2.86, 6.00, 1.85, 1.41],
    'Grau de Endividamento': [43.16, 76.33, 52.58, 20.81, 47.38, 44.76],
    'Margem Líquida': [59.48, 12.59, 19.22, 26.90, 20.10, 48.59],
    'ROE': [10.56, 19.73, 14.49, 7.18, 5.76, 16.72],
    'P/L': [11.50, 11.47, 5.90, 9.63, 152.97, 13.67],
    'P/VPA': [0.71, 11.47, 0.80, 0.63, 1.53, 2.09],
    'Dividend Yield': [5.39, 3.08, 5.35, 4.91, 2.18, 3.91]
}

df = pd.DataFrame(dados)
```

```python
# Normalização
maior_melhor = ['Liquidez Corrente', 'Margem Líquida', 'ROE', 'Dividend Yield']
menor_melhor = ['Grau de Endividamento', 'P/L', 'P/VPA']

for col in maior_melhor:
    df[col + '_norm'] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())

for col in menor_melhor:
    df[col + '_norm'] = 1 - (df[col] - df[col].min()) / (df[col].max() - df[col].min())

pesos = {
    'Liquidez Corrente_norm': 0.10,
    'Grau de Endividamento_norm': 0.15,
    'Margem Líquida_norm': 0.15,
    'ROE_norm': 0.15,
    'P/L_norm': 0.15,
    'P/VPA_norm': 0.10,
    'Dividend Yield_norm': 0.10
}

# Cálculo do Score Final
df['Score Final'] = sum(df[col] * peso for col, peso in pesos.items())
df_ranked = df[['Empresa', 'Score Final']].sort_values(by='Score Final', ascending=False).reset_index(drop=True)
```

```python
# gráfico 
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(df_ranked['Empresa'], df_ranked['Score Final'], color='#66b3ff', edgecolor='black')
ax.set_title('Ranking Final das Empresas (Score Normalizado)', fontsize=16, fontweight='bold')
ax.set_xlabel('Empresa', fontsize=14)
ax.set_ylabel('Score Final', fontsize=14)
ax.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
